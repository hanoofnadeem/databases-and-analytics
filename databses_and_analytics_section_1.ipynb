```r
library(sqldf)


ideas_per_year <- sqldf('
  SELECT
    substr("Idea.Submission.Date", 7, 4) AS Year,
    COUNT(*) AS NumberOfIdeas
  FROM data
  GROUP BY Year
  ORDER BY Year
')

print(ideas_per_year)
```

```r
security_flag_by_office <- sqldf("
  SELECT
    `Office.Location`,
    COUNT(*) AS Total_Ideas,
    SUM(CASE WHEN `Security.Concern.Flag` = 'Yes' THEN 1 ELSE 0 END) AS Security_Flagged,
    ROUND(100.0 * SUM(CASE WHEN `Security.Concern.Flag` = 'Yes' THEN 1 ELSE 0 END) / COUNT(*), 2) AS Security_Flag_Percentage
  FROM data
  GROUP BY `Office.Location`
  ORDER BY Security_Flag_Percentage DESC
")

print(security_flag_by_office)
```

```r
# Load necessary packages
library(tidyr)  # For pivot_wider function to reshape data
library(dplyr)  # For data manipulation functions like group_by and summarise

# Summarise and reshape data to show idea counts by Department and Approval Status
approval_summary <- greenfuture_ideas_dataset %>%
  # Group data by Department and Approval.Status columns
  group_by(Department, Approval.Status) %>%

  # Count the number of ideas in each group
  summarise(Idea_Count = n()) %>%

  # Reshape the data: turn Approval.Status values into column names (e.g. Approved, Rejected)
  # and fill missing combinations with 0
  pivot_wider(
    names_from = Approval.Status,      # Values in Approval.Status become column names
    values_from = Idea_Count,          # Values to fill those columns come from Idea_Count
    values_fill = 0                    # If no ideas for a category, fill with zero instead of NA
  )

# Display the reshaped summary dataframe
approval_summary
```

```r
library(ggplot2)  # Load ggplot2 for plotting

# Create a boxplot to show the distribution of votes in each department
ggplot(greenfuture_ideas_dataset, aes(x = Department, y = Number.of.Votes)) +
  geom_boxplot(fill = "skyblue", color = "darkblue") +  # Add boxplot with custom colors
  theme_minimal() +  # Apply a clean, minimal theme
  labs(title = "Vote Distribution by Department",  # Title of the plot
       x = "Department",  # X-axis label
       y = "Number of Votes")  # Y-axis label
```

```r
library(dplyr)    # For data manipulation
library(tidyr)    # For pivoting data
library(ggplot2)  # For plotting

# Step 1: Group and count ideas by Department and Idea Category
idea_counts <- greenfuture_ideas_dataset %>%
  group_by(Department, Idea.Category) %>%
  summarise(Count = n(), .groups = "drop")  # Count ideas and remove grouping

# Step 2: Pivot the data to wide format for heatmap
idea_matrix <- idea_counts %>%
  pivot_wider(names_from = Idea.Category, values_from = Count, values_fill = 0)

# Step 3: Convert to matrix format for heatmap
heatmap_data <- as.data.frame(idea_matrix)
rownames(heatmap_data) <- heatmap_data$Department
heatmap_matrix <- as.matrix(heatmap_data[,-1])  # Remove department column

# Step 4: Plot heatmap
heatmap(heatmap_matrix,
        Rowv = NA, Colv = NA,  # No clustering
        col = heat.colors(256),  # Color palette
        scale = "column",        # Normalize column-wise
        margins = c(6, 10),      # Adjust margins
        xlab = "Idea Category",
        ylab = "Department",
        main = "Heatmap of Idea Counts by Department and Category")
```

```r
# Load necessary libraries
library(dplyr)      # For data manipulation
library(ggplot2)    # For data visualization
library(lubridate)  # For date formatting

# Step 1: Convert 'Idea.Submission.Date' to Date format and extract 'Month' as "YYYY-MM"
greenfuture_ideas_dataset <- greenfuture_ideas_dataset %>%
  mutate(Submission.Date = as.Date(Idea.Submission.Date),              # Convert to Date format
         Month = format(Submission.Date, "%Y-%m"))                     # Extract Year-Month

# Step 2: Group data by 'Month' and count the number of ideas submitted in each month
monthly_submissions <- greenfuture_ideas_dataset %>%
  group_by(Month) %>%
  summarise(Idea_Count = n(), .groups = "drop")                        # Count ideas per month

# Step 3: Create a line plot showing trend of idea submissions over time
ggplot(monthly_submissions, aes(x = Month, y = Idea_Count, group = 1)) +
  geom_line(color = "steelblue", size = 1.2) +                         # Add a line in blue
  geom_point(color = "darkred", size = 2) +                            # Add data points
  labs(
    title = "Monthly Idea Submissions Over Time",                     # Title
    x = "Month",                                                      # X-axis label
    y = "Number of Ideas Submitted"                                   # Y-axis label
  ) +
  theme_minimal(base_size = 14) +                                     # Use a clean minimal theme with larger text
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1),                # Rotate x-axis text for readability
    plot.title = element_text(face = "bold", hjust = 0.5)             # Center and bold the title
  ) +
  scale_x_discrete(guide = guide_axis(n.dodge = 2))                   # Avoid x-axis label overlapping
```

```r
# Load required package
library(ggplot2)

# Convert Month to a factor for better control over axis order (optional)
greenfuture_ideas_dataset$Month <- factor(greenfuture_ideas_dataset$Month, levels = unique(greenfuture_ideas_dataset$Month))

# Create scatter plot
ggplot(greenfuture_ideas_dataset, aes(x = Month, y = Number.of.Votes)) +
  geom_point(color = "steelblue", size = 3, alpha = 0.6) +  # points with slight transparency
  labs(
    title = "Scatter Plot of Votes Across Months",
    x = "Month",
    y = "Number of Votes"
  ) +
  theme_minimal(base_size = 13) +  # cleaner theme
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1),  # rotate x labels
    plot.title = element_text(hjust = 0.5, face = "bold")
  )
```

```r
# Load the ggplot2 package for visualization
library(ggplot2)

# Generate a density plot to show the distribution of number of votes
ggplot(greenfuture_ideas_dataset, aes(x = Number.of.Votes)) +

  # Add a smooth density curve with orange fill and some transparency
  geom_density(fill = "darkorange", alpha = 0.6) +

  # Add axis labels and a plot title
  labs(
    title = "Density Plot of Number of Votes",  # main title of the plot
    x = "Number of Votes",                      # x-axis label
    y = "Density"                               # y-axis label
  ) +

  # minimalistic theme for better readability
  theme_minimal(base_size = 13) +

  #  plot title
  theme(
    plot.title = element_text(hjust = 0.5, face = "bold")
  )
```
